## Unités de mémoire
Nous avons vu qu'un ordinateur utilise le binaire. La plus petite unité de mémoire pour un ordinateur est le bit. Un bit peut avoir pour valeur 0 ou 1.

> Euh, je croyais que la mémoire d'un ordinateur se comptait en gigas.

Giga n'est pas une unité, mais un préfixe, comme kilo, méga, milli, ... Il faut savoir quelle unité on met derrière : gigagramme ? gigaseconde ? gigamètre ?

En informatique, l'unité de mémoire est l'octet. Un octet est un groupe de 8 bits (le préfixe oct correspond à 8). En anglais, on utilise le terme byte (littéralement, une bouchée). Cela vient du nombre de bits que pouvaient "manger" les premiers processeurs. Au début, cela pouvait varier d'un constructeur à l'autre. Avec l'informatique grand public, les premiers ordinateurs étaient construits autour de processeur 8 bits. Le byte s'est donc stabilisé à 8 bits et nous appelons cela un octet.

Comme toute unité, on a des abbréviations:

Unité | Abbréviation
--- | ---
octet | o
byte | B
bit | b

Certains fabricants jouent parfois sur l'ambiguïté entre byte et bit (surtout que leurs abbréviations ne diffèrent que par la casse). Ainsi, la Playstation 1 utilisait des cartes mémoires de 128 ko. Mais certains vendeurs proposaient des cartes de 1Mo. En réalité, il était écrit sur la carte : 1 Mb, or 1 Mb = 1024 kb = 1024 / 8 kB = 128 kB (ou ko si vous préférez). Bref, on était tout content d'acheter une carte 8 fois plus grande que la normale pour seulement 2 fois plus cher. Et puis, on se retrouvait avec une carte ordinaire...

> Euh, m'sieur, il y a une erreur : 1 Mb = 1000 kb (je l'sais, j'ai vu ça en physique).

Oui, en physique, les préfixes correspondent à des puissances de 10 :

Préfixe | Abbréviation | Nombre d'unités
--- | --- | ---
kilo | k | $10^{3}$
Méga | M | $10^{6}$
Giga | G | $10^{9}$
Téra | T | $10^{12}$
Péta | P | $10^{15}$
Exa  | E | $10^{18}$

Un préfixe correspond donc à 1000 ($10^{3}$) fois le préfixe précédent.

Oui, mais en informatique, on compte en binaire. Tout tourne donc autour de puissances de 2. Heureusement, il y a une puissance de 2 qui se rapproche de $10^{3}$, c'est $2^{10} = 1024$.

On a donc créé de nouveaux préfixes pour l'informatique : 

Préfixe | Abbréviation | Nombre d'unités
--- | --- | ---
kiloinformatique | ki | $2^{10} = 1024$
Mégainformatique | Mi | $2^{20} = 1024^2$
Gigainformatique | Gi | $2^{30} = 1024^3$
Térainformatique | Ti | $2^{40} = 1024^4$
Pétainformatique | Pi | $2^{50} = 1024^5$
Exainformatique  | Ei | $2^{60} = 1024^6$

Dans l'exemple de la carte mémoire ps1, il aurait fallu écrire : 1 Mib = 1024 kib pour être exact.

Là encore, des fabricants jouent sur l'ambiguïté : on achète une clé de 128 Go, et notre ordinateur nous dit qu'elle fait seulement 119 Go. Le fabricant n'a pas menti. Sa clé fait bien $128 Go = 128 \times 10^9 o$, mais votre ordinateur compte en Gio (même si Windows écrit Go). 

Or $128 \times 10^9 o = \frac{128 \times 10^9}{1024} kio = 125 \times 10^6 kio = \frac{125 \times 10^6}{1024} Mio = 122070 Mio = \frac{122070}{1024} Gio = 119 Gio$

Et tout ça, sans compter la place occupée par la table des matières qui n'est pas disponible pour stocker le contenu des fichiers mais que le fabricant peut compter dans la taille mémoire qu'il vous vend. Et ouais...

On va chercher un encadrement avec des puissances de 2 : 

$2^{17} = 131072$ donc avec 17 bits, on peut au maximum coder 131071. Ce n'est pas assez pour coder 204875.

Cherchez combien de bits sont nécessaires pour coder :
- 594
- 800
- 900
- 10084

Ecrivez un programme qui calcule le nombre de bits nécessaire au codage d'un nombre :

````python
def nbBits(nombre):
    n = 0
    while 2**n < nombre:
        n += 1
    return n

print(nbBits(594))   #10
print(nbBits(800))   #10
print(nbBits(900))   #10
print(nbBits(10084)) #14
````

## Opérations logiques binaires
Le binaire permet de nouvelles opérations. Rassurez-vous, nous en avons plus ou moins déjà parlé.

### & : le ET logique
Vous vous souvenez des instructions conditionnelles. Quand on veut voir si deux conditions sont vraies en même temps, on utilise `and`. Ici, nous allons parler de l'opérateur binaire `&`. Ce n'est pas *tout à fait* la même chose.

Comme pour l'addition et la multiplication, on peut établir la table de cette opération.

& | 0 | 1
--- | --- | ---
0 | 0 | 0
1 | 0 | 1

Pour mieux comprendre comment on l'obtient, pensez à la phrase : "Il fait beau et chaud". On y trouve deux conditions: 
- Il fait beau.
- Il fait chaud.

Chaque condition peut être vraie (1) ou fausse (0).

Si le ciel est bleu et sans nuage, la condition "il fait beau" est vérifiée, mais il est possible que la température soit de 10°C et on ne peut pas dire qu'il fait chaud. La seconde condition est donc fausse. La phrase "Il fait beau et chaud" est donc fausse et le résultat de l'opération "et" est 0.

De même, il peut faire 30°C et donc il fait chaud, mais le ciel peut être couvert et il peut pleuvoir et il ne fait donc pas beau. Encore une fois, il est faux de dire qu'il fait beau et chaud.

Pour que cette phrase soit vraie, il faut donc que chacune des deux conditions soit vraie.

Bien sûr, s'il fait 10°C et qu'il pleut, la phrase est fausse.

Il faut faire la différence entre `and` et `&`. Quand on utilise deux nombres de plus d'un bit, l'opérateur `&` réalise un ET logique entre le bits de chaque nombre rang par rang :
```
n1      :  101101011   (363)
n2      :  000110111   (55)
n1 & n2 :  000100011   (35)
```
Dans le cas de `and`, si on l'utilise entre deux nombres, alors si l'un des deux nombre vaut 0, le résultat sera 0. Sinon, ce sera le deuxième nombre (en Python). 

`363 & 55` donne donc 35 tout comme `55 & 363`, mais `363 and 55` donne 55 et `55 and 363` donne 363.

D'une manière générale, `and` est à utiliser entre deux conditions, alors que `&` sert dans des calculs entre nombres.

> Mais à quoi ça sert ?

A plein de choses. Par exemple, une adresse IP & un masque de sous-réseau donne l'adresse du réseau. Mais si, vous l'avez vu en SNT l'an dernier, vous ne l'avez pas dit comme ça, c'est tout.

### | : le OU logique
Comme pour `&`, il existe un OU logique `|`.

Pour trouver sa table, pensez à "Il faut beau **ou** chaud" :
- S'il fait 10 °C et que le ciel est bleu, c'est vrai.
- S'il fait 30 °C mais qu'il pleut, c'est vrai.
- S'il fait 30 °C et que le ciel est bleu, c'est vrai
- S'il fait 10 °C et qu'il pleut, c'est faux.

Voici donc la table du OU :

\| | 0 | 1
--- | --- | ---
0 | 0 | 1
1 | 1 | 1

```
n1      :  101101011   (363)
n2      :  000110111   (55)
n1 & n2 :  101111111   (383)
```

Là encore, il ne faut pas le confondre avec `or`.

### ^ : le OU exclusif
Vous vous êtes peut-être demandé pourquoi, en Python, la puissance n'était pas représentée par `^` mais par `**`. Dans d'autres langages, on a même recours à une fonction.

Et bien, c'est parce que l'opérateur `^` est utilisé pour le OU exclusif (en anglais: "eXclusive OR" que l'on abrège en XOR).

Pour trouver sa table, pensez à "Il faut beau ou chaud, **mais pas les deux en même temps**."

^ | 0 | 1
--- | --- | ---
0 | 0 | 1
1 | 1 | 0

```
n1      :  101101011   (363)
n2      :  000110111   (55)
n1 & n2 :  101011100   (348)
```

### ~ : le NON
Lorsque nous avons parlé des conditions, nous avons vu l'opérateur `not` qui transforme `True` en `False` et vice-versa.

Il existe un opérateur qui fait la même chose au niveau des bits d'un nombre. Sa table est la suivante :

bit  | 0 | 1
--- | --- | ---
~ | 1 | 0

```
n      :  101101011   (363)
~n     :  010010100   (148 ?)
```
